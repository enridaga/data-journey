```python
# Group data by Dataset/InputSize/TestRegime/Method and compute AVG & STD on each Measures
# avg = data[(data['Dataset']=='BC_VisualisationPreparation.csv') & (data['InputSize']==10) & (data['TestRegime']==1) & (data['Method']=='Decision Tree')].mean()
# std = data[(data['Dataset']=='BC_VisualisationPreparation.csv') & (data['InputSize']==10) & (data['TestRegime']==1) & (data['Method']=='Decision Tree')].std()
# avg, std
# # https://stackoverflow.com/questions/36106490/how-to-get-unique-values-from-multiple-columns-in-a-pandas-groupby
```

```python
def Table(dataframe, dataset, test_regime):
    # VP, TestRegime 1
    df = dataframe[dataframe['Dataset'] == dataset].drop(columns=['Notebooks','Dataset'])
    TR = df.loc[df['TestRegime']==test_regime].drop(columns=['TestRegime'])
    return TR.groupby(['InputSize','Method']).agg(['mean','std'])
```

```python
Table(data, 'MultiClassification.csv', 3 )
```
